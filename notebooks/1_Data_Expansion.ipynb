```python
import pandas as pd

# Load datasets
load_df = pd.read_csv(r"C:\\Users\\reddy\\OneDrive\\Documents\\TSA-PROJECT\\data\\processed\\load_data_2022_final.csv")
weather_df = pd.read_csv(r"C:\\Users\\reddy\\OneDrive\\Documents\\TSA-PROJECT\\data\\raw\\weather_data.csv")

# Convert date columns
load_df['date'] = pd.to_datetime(load_df['date'])
weather_df['date'] = pd.to_datetime(weather_df['date'])

print(load_df.shape, weather_df.shape)
load_df.head()
```

```python
#merge on date
df = pd.merge(load_df, weather_df, on='date', how='inner')

print("Merged shape:", df.shape)
df.head()
```

```python
#check missing values
print(df.isnull().sum())
```

```python
#save multivariate dataset
OUTPUT_FILE=r"C:\\Users\\reddy\\OneDrive\\Documents\\TSA-PROJECT\\data\\processed\\load_weather_2022.csv"

df.to_csv(OUTPUT_FILE,index=False)
print("Merged dataset saved at:",OUTPUT_FILE)
```
